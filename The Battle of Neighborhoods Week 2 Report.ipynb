# 1. Introdunction

## 1.1 Background

The United States Tennis Association (USTA) is planning to organize a number of Tennis tournaments, coaching courses, seminars and workshops for professional tennis players and coaches from all over the world during the next few months. All the events mentioned above are scheduled to take place in The National Tennis Developement Center located in Flushing Meadows, New York. Most of the players and coaches participating in any event will require lodging and dining facilities throughout the duration of the event. Additionally, the management of USTA has decided that a gymming facility shall also be made available to the participants. 

## 1.2 Problem

The National Tennis Developement Center does not have any facilities for the lodging, dining or gymming of the participants. So, after a number of meetings, the management had decided to collaborate with a few Hotels, Restaurants and Gyms within a specified radius from the academy in order to provide the participants with the required facilities. The Data Science Department of the USTA had been tasked with fulfilling the following objectives:
    
   1. Identify 8 Hotels that are nearest to the academy within a 1.5 KM radius.
   2. Identify 8 Restaurants that are nearest to the academy within a 500 meter radius.
   3. Identify 3 Gyms that are nearest to the academy within a 3 KM radius.
   4. Visualize the locations identified on one single map.

There is a requirement for 5 Hotels, 5 Restaurants and 2 Gyms. So, the locations identified by the Data Science Department will be scrutinized and explored further by the management for the selection of suitable Hotels, Restaurants and Gyms to collaborate with.

## 1.3 Interest

The management is interested in getting information about the nearby Hotels, Restaurants and Gyms as this information will definately help in making timely arrangements and the events will prove to be a good experience for the participants. Additionally, information about the lodging, dining and gyming facilities to be provided may also be forwarded to the participating players and coaches before they arrive. They will be interested in this information as it will help them in planning and scheduling their trip accordingly.

# 2. Data

## 2.1 Data Aquisation

The data used in this project is based on Foursquare location data for Flushing, NY 11368, United States. It was acquired using Foursquare's API. Using Foursquare location data, it was possible to explore all the Hotels, Restaurants and Gyms within the specified radius of the venue. 

## 2.2 Data Cleaning

The data was obtained from Foursquare's API in the JSON format. The relevant part of the JSON object were converted into a DataFrame. The rows with missing values were dropped from the DataFrame.

## 2.3 Feature Selection

Data related to the locations of interest was extracted using Foursquare and it was used to produce DataFrames. The data used in producing the DataFrames is as follows:
    
   1. Name of the location of interest.
   2. Category of the location of interest.
   3. Address of the location of interest.
   4. Distance of the location of interest from the venue.
   5. Latitude of the location of interest.
   6. Longitude of the location of interest.
   7. Postal Code of the location of interest.

# 3. Methodology

In this project, the efforts were directed on identifying the three types of locations highlighted in the Introduction Section, namely, Hotels, Restaurants and Gyms. In particular, those locations were selected that were within the specified radius of the The National Tennis Developement Center.

## 3.1 Exploratory Data Analysis

1. As the first step, all the necessary libraries for obtaining the data, it's exploration, analysis and visualization were imported.

2. The second step was to define the Foursquare credentials.

3. As the third step, the coordinates of the venue were obtained.

4. Next, the following steps were performed for each of the three locations of interest (Hotels, Restaurants and Gyms):

   - A URL for the Foursquare API was defined.
   - The data related to the location of interest was obtained using the Foursquare API in JSON format.
   - The relevant part of the JSON object was converted into a DataFrame
   - The DataFrame as cleaned and all the rows containing NaN or Null values were dropped. Additionally, all the useless columns were also dropped.
   - The rows of the DataFrame were sorted in ascending order of the 'distance' column and the required number of rows were selected from the top.
   

5. In the final step, the data in the DataFrames was used to visualize the locations on a map centered around the venue. This was done using the Folium library.

# 5. Discussion

The analysis shows that there are a lot of Hotels and Restaurants around the venue but there are not a lot of Gyms. The density of Hotels and Restaurants around the venue is also very good but the Gyms are quite scattered. Attention was directed towards the Hotels, Restaurants and Gyms that are located within the specified radius from the venue. Then, the location data was cleaned and organised. The missing values were dealt with and the useless data was dropped. Next, the Hotels, Restaurants and Gyms were sorted in ascdending order of their distance from the venue. After that, the required number of Hotels, Restaurants and Gyms were selected. Finally, these selected locations of interest were visualized on a map centered around the venue. # 6. Conclusion

The purpose of this project was to identify 8 nearest Hotels within 1.5 kilometers, 8 nearest Restaurants within 500 metres and 3 nearest Gyms within 3 kilometers of Flushing, NY 11368, United States. This project was undertaken in order to aid the management of USTA in finding suitable Hotels, Restaurants and Gyms to collaborate with. By using location data for Hotels, Restaurants and Gyms from Foursquare, the locations were identified, sorted and selected based on the requirements put forward by the management.

The final decision on the suitable Hotels, Restaurants and Gyms for collaboration will be made by the management based on specific characteristics of each location, taking into consideration additional factors like facilities offered, affordability, reputation, approach from the venue, connectivity from the major parts of the city, criminal records, administration, etc.
